# Bias vs Variance

The ML models depend on many factors, one of them is the hyper-parameters. How should we choose our hyper-parameters so that we can be sure that we are getting the best results? 

The answer to this question may come from the equilibrium between bias and the variance. A high biased model is not able to predict neither the training and the testing sets, its complexity is not high enough to predict anything. On the other hand, a high varianced model predicts perfectly the training set but it can not predict the testing set, it has just "memorized" the training set and it can not make any prediction out of that set. A biased model underfits the data, a high varianced model is an overfited one. 

 ![picture](https://github.com/jakevdp/PythonDataScienceHandbook/blob/master/notebooks/figures/05.03-validation-curve.png?raw=true)

The validation score (a $R^2$ for example) for a training set will tend to a maximum that works like an asymptote, no mater how we increase the complexity of the model, the training score will not improve. And for the validation score we will have a point where it gets a maximum and after that the performance of the model can not produce a better result. The point where the validation score starts to drop show us the maximal complexity for the model.

```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn import metrics

from sklearn.model_selection import validation_curve
from sklearn.model_selection import learning_curve
from sklearn.utils import resample

import matplotlib.pyplot as plt
```

```python
def polyData(n, seed=1, err=1):
    if seed:np.random.seed(seed)
    xs = (1-0.1)*np.random.rand(n)+0.1
    ys = 10-1/xs + err*np.random.normal(loc=0.0, scale=1.0, size=n)
    return xs.reshape(-1,1), ys
```

```python
x, y = polyData(100, seed=16, err=1)

plt.figure(figsize=(15,8))
plt.scatter(x,y, c="k", alpha=0.5)
```

```python
#and now lets make a function that creats a polynomical model of degree n to see how it fits the data.
#For this we use a pipeline
def PolynomicalModel(degree=2, **kwards):
    return make_pipeline(PolynomialFeatures(degree), 
                         LinearRegression(**kwards))
```

# Cross Validation $\to$ validation_curve()

```python
#validation_curve() computes the scores of training and testing sets for different values of a parameter of the 
#estimator. It performs a cross validation process with cv divisions. So it returns a tuple with two arrays;
#the first one with the scores of the training set and the other one for the testing sets. Each of them with 
#a length of len(param_range). The lists inside the arrays have cv elements corresponding to each training/testing
#score in the cross validation process.
degs = np.arange(0,21,1)
training_score, validation_score = validation_curve(PolynomicalModel(), 
                                                    X=x, y=y, cv=5, 
                                                    param_name="polynomialfeatures__degree", param_range=degs)
PolynomicalModel().get_params()
```

```python
plt.figure(figsize=(10,7))
plt.plot(degs, training_score.mean(axis=1),   marker="o", label="Training score")
plt.plot(degs, validation_score.mean(axis=1), marker="s", label="Testing score")
plt.xlim(min(degs), max(degs))
plt.xlabel("Polynomial degree (Complexity)")
plt.ylabel("Score $R^2$")
plt.xticks(ticks=degs, labels=degs)
plt.legend(frameon=False, loc="lower right")
```

we see that the testing score is decreasing as we increase the complexity of the model but the training score
tends asymptotically to a value (~0.8). The value where the testing score starts to decrease is the 
optimum complexity of the model, in this case may be a 3rd degree polynomial. 

```python
#the validation curve depends not only on the complexity oof the model but also on the number of points
#it has:
x2, y2 = polyData(500, seed=235, err=1)
training_score2, validation_score2 = validation_curve(PolynomicalModel(), 
                                                    X=x2, y=y2, cv=5, 
                                                    param_name="polynomialfeatures__degree", param_range=degs)

plt.figure(figsize=(10,7))
#500 points
plt.plot(degs, training_score2.mean(axis=1),   marker="o", linestyle="-.", label=f"Training score {len(x2)}")
plt.plot(degs, validation_score2.mean(axis=1), marker="s", linestyle="-.", label=f"Testing score {len(x2)}")

#100 points
plt.plot(degs, training_score.mean(axis=1),   marker="o", label=f"Training score {len(x)}")
plt.plot(degs, validation_score.mean(axis=1), marker="s", label=f"Testing score {len(x)}")

plt.xlim(min(degs), max(degs))
plt.xlabel("Polynomial degree (Complexity)")
plt.ylabel("Score ($R^2$)")
plt.xticks(ticks=degs, labels=degs)
plt.legend(frameon=False, loc="lower right")
plt.grid()
```

If we have more data our model could need (and work and work better with) a more complex model. 
So we see that the score not only depends on the complexity of the model but also on the size of the 
training set. To measure the dependency of the model on the training size we can use the learning_curve method
from sklearn.

validation_curve give us an array with the scores for each value of the complexity parameter for all the cv.
learning_curve Determines cross-validated training and test scores for different training set sizes.
give us an array with the scores for each sample size in the training set for all the cv
performed for the  given complexity parameter. It also returns the sizes of the training sets for which it 
was perfromed the scores. 

```python
#lets see how the size of the training set affects the score for different levels of complexity (degrees)
degrees = [3, 6, 9, 12, 15, 18]
sizes   = np.linspace(0.1,1,20)

columns = 3
rows    = int(len(degrees)/columns) if len(degrees)%columns==0 else len(degrees)//columns+1

plt.figure(figsize=(7*columns,7*rows))
i = 1
for degree in degrees:
    #generating the training/testing scores for each degree
    N, r2_training, r2_testing = learning_curve(estimator=PolynomicalModel(degree), 
                                                X=x2, y=y2, cv=8, 
                                                train_sizes=sizes)
    #plot
    plt.subplot(rows, columns, i); i+=1
    plt.plot(N, np.mean(r2_training, axis=1), label=r"Training $R^2$")
    plt.plot(N, np.mean(r2_testing, axis=1),  label=r"Testing  $R^2$")
    plt.hlines(np.mean([max(np.mean(r2_testing, axis=1)), min(np.mean(r2_training, axis=1))]), N[0], N[-1], "k")
    
    plt.title(f"Degree = {degree}")
    plt.xlabel("Number of points")
    plt.ylabel(r"$R^2$")
    plt.ylim(0.7, max(np.mean(r2_training, axis=1)))
    plt.xlim(N[0], N[-1])
    plt.legend(frameon=False)
    plt.grid()
```

# Bootstrap

```python
#Here we have a nice explanation about the differences between CV and Bootstrap
#https://medium.com/codex/two-resampling-approaches-to-assess-a-model-cross-validation-and-bootstrap-794916d37212
```

The k-folds Cross Validation splits the data in k subsets and takes k-1 of them for the training set and performs this procedure k times, each leaving out of the training set a diferent subset of the k. In this way it give us the scores for each one of the k-folds for each model (with the param_range argument we give the models).

Bootstrap makes subsets of a given size choosing randomly the elements. We can choose either if we want a same element to be able to repeted or not. With this procedure we can create several training sets and for each of then get their statictics, in this way it allows us to give confidential intervals for the statistics we are using. 

```python
data = np.array([x2.reshape((1,-1))[0],y2]).T
n = 1000
training_size = 0.8

metric = []

for i in range(n):
    #generaiting the training (bootstrap) and testing (oob (oob->out of the bag)) sets
    bootstrap = resample(data, n_samples=int(len(data)*training_size), replace=True)
    oob       = np.array([x for x in data if x not in bootstrap])
    
    #the training and testing sets
    x_train = bootstrap[:,0].reshape((-1,1))
    y_train = bootstrap[:,1]
    x_test  = oob[:,0].reshape((-1,1))
    y_test  = oob[:,1]
    
    #training the model and predicting
    poli_model = PolynomicalModel(degree=4)
    poli_model.fit(x_train, y_train)
    y_pred = poli_model.predict(x_test)
    
    #metrics
    #metric.append(metrics.mean_absolute_error(y_pred=y_pred, y_true=y_test)**0.5)
    metric.append(poli_model.score(X=x_test, y=y_test))

#plotting the metrics for the model
plt.figure(figsize=(5,5))
hist = plt.hist(metric, bins=int(np.sqrt(len(RMSE))), histtype="step")
plt.vlines(np.mean(metric), 0, max(hist[0])*(1+0.1), color="k" )
plt.fill_betweenx(y=np.linspace(0,max(hist[0])*(1+0.1)), 
                  x1=np.mean(metric)-np.std(metric), x2=np.mean(metric)+np.std(metric), 
                  color="orange", alpha=0.5)
plt.ylim(0, max(hist[0])*(1+0.1))
plt.grid()
print(f"average = {np.mean(metric)} +/- {np.std(metric)}")
```
