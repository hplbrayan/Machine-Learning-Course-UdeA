Los [drivers](https://github.com/mkleehammer/pyodbc/wiki/Connecting-to-SQL-Server-from-Windows) hace referencias a los tipo de conexión para consumir los datos, de este tipo de conexiones importante conocer dos:

* Open Database Connectivity [ODBC](https://es.wikipedia.org/wiki/Open_Database_Connectivity#:~:text=Open%20DataBase%20Connectivity%20(ODBC)%20es,(DBMS)%20almacene%20los%20datos.)
* Java Database Connectivity [JDBC]( https://es.wikipedia.org/wiki/Java_Database_Connectivity#:~:text=Java%20Database%20Connectivity%20(en%20espa%C3%B1ol,datos%20a%20la%20cual%20se)

Server hace referencia a el nombre del servidor SQL (también puede ser localhost si se está conectando localmente) y Database hace referencia al nombre de la base de datos. Dado esos valores identificadores tenermos dos opciones de conexión:

````
#Permisos por una conexión segura de Windows
connection_str_windows_auth = (
  r'Driver={ODBC Driver 17 for SQL Server};'
  r'Server=localhost;'
  r'Database=Testing;'
  r'Trusted_Connection=yes;'
  )

#Permisos por unas credenciales User/Password
connection_str_sql_user = (
  r'Driver={ODBC Driver 17 for SQL Server};'  
  r'Server=localhost/IP/SERVER;'
  r'Database=DATABASE;' 
  r'UID=USERNAME;' 
  r'PWD=PASSWORD;' 
  )
````

```python
try:
  import pyodbc
  import sqlalchemy as sa
  import pandas as pd
  #Crear un objeto conector
  conn = pyodbc.connect(connection_str_windows_auth)
  cursor = conn.cursor()

  #Llamar una tabla
  query = "SELECT * FROM Pruebas.TablaVacia"
  test = pd.read_sql(query, conn)
  print(test)

  #Procesar
  test['coladd'] = test['colvarchar'] + 'ADD'
  print(test)

  #Vamos a escribir la tabla en nuestra base de datos
  connStr = 'mssql+pyodbc://@localhost/Testing?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
  engine = sa.create_engine(connStr)
  test.to_sql('TablaADD',schema="Pruebas",con=engine)

except:
  print("No local")
```
