```python
#We read a JSON from datos.gov.co
#we add the "limit" append to the path to obtain the whole data set since by default it wont allow us 
#to download more than 1000 rows, this to avoid "breaking" the page
#path = "https://www.kaggle.com/harikrishna9/hotel-feed-back-data?select=Sample+hotel+data.csv"
path = "https://www.datos.gov.co/resource/kktn-hnps.json"
limit = "?$limit=100"
data = pd.read_json(path+limit)
```
