# Series

The series in pandas are one dimensional arrays, they can be thought as the colums of a DataFrame

```python
a = [1, 2, 3]
a = pd.Series(a)
print(a)

#to obtain elements
a[1]
```

```python
#by default pandas will give as index for each element their position
#but if we want we can give them our own indexes by using the index argument:
a = [1, 2, 3]
myIndex = ["a", "b", "c"]
a = pd.Series(a, index=myIndex)
print(a)

#and use them to get its value:
a["b"]
```

```python
#or we can create series from a dic
a = {"a":4, "b":5, "c":6}
a = pd.Series(a)
print(a)

#we can even create the series just by using some of the indexes
a = pd.Series(a, index=["a", "c"])
a
```

# DataFrame

The DataFrames are n-dimentional, so they can contain several columns. Each individual column in
a DataFrame is a Series. 

```python
#To create a DataFrame we use
a = {"a": [1,2,3], "b":[4,5,6], "c":[7,8,9]}
a = pd.DataFrame(a)
a
```

```python
#if we try to convert into a series we would have a one dimensional array:
aS = {"a": [1,2,3], "b":[4,5,6], "c":[7,8,9]}
aS = pd.Series(aS)
aS
```

```python
#So here we see a big difference between both, Series and DataFrame objects in pandas.
#We can use the loc attribute to return the values from one or more columns:
print(a.loc[0])
print("\n")
print(a.loc[[0,2], ["a", "c"]])
```

```python
#and as we can do with series we can also give custum indexes for our rows:
a = {"a": [1,2,3], "b":[4,5,6], "c":[7,8,9]}
myIndex = ["row 1", "row 2", "row 3"]

a = pd.DataFrame(a, index=myIndex)
a
```

# 1. Empty cells

```python
#If we have some data and it contains empty cells this may cause errors in our calculations
#so we want to do something with these empty cells, an option is to delete them, since 
#we usually have lots of rows deleiting some of them is not going to be so important
#for this we use the dropna() method on the DataFrame
new_data = data.dropna()
new_data
```

```python
new_data.info()
```

```python
#in this new data we can see that there are not null values.
#When using dropna() it creates a new DataFrame, if we want to the original one to be 
#modifided instead of create a new one we can use the argument inplace=True:
data.dropna(inplace=True)
data
```

```python
data.info()
```

```python
#but sometimes we do not want to delete the rows with empty values, we can also fill this empty 
#values with another value by using the fillna() method
data = pd.read_csv("https://www.w3schools.com/python/pandas/dirtydata.csv.txt")
data.fillna(99999, inplace=True)
data
```

```python
#but using fillna() in this way will change every null value in the whole DataFrame, if we want 
#just to change the values for som columns we can do it like this:
data = pd.read_csv("https://www.w3schools.com/python/pandas/dirtydata.csv.txt")
data["Calories"].fillna(99999, inplace=True)
data
```

```python
#or we can use the mean, the mode, the median or any other value to fill the columns
data = pd.read_csv("https://www.w3schools.com/python/pandas/dirtydata.csv.txt")
mean = data["Calories"].mean()
data["Calories"].fillna(mean, inplace=True)
print(mean)
data
```

# df.apply()

```python
#an useful method to perform operation over columns and rows is apply()
data.Calories.apply(lambda x:x*2+12)
```

```python
#we can choose to what axis to apply the function
#axis=0 works for the columns (default)
data.apply(np.sum, axis=0)
```

```python
#axis=1 operates over the rows
data.apply(np.sum, axis=1)
```

```python
#apply has the argument result_type
#the argument result_type with "broadcast" will broadcast the reult of applying the function to the whole column if axis=0
#and to the whole row if axis=1
data.apply(np.sum, axis=0, result_type="broadcast")
```

```python
data.apply(np.sum, axis=1, result_type="broadcast")
```

```python
#result_type="expand" allow us to add some new rows
#we cand add the new columns to the original data frame
data.apply(lambda cell: [cell.Duration*2, cell.Pulse+10], axis=1, result_type="expand")
```

```python
#if insted result_type="expand" we use result_type="reduce" it will return a new series
data.apply(lambda cell: [cell.Duration*2, cell.Pulse+10], axis=1, result_type="reduce")
```

```python
#the method applymap() applyes a function to the whol data frame, sometimes works fasrter than apply
data.applymap(lambda cell: cell+1)
```

# 2. Data with wrong format

```python
#In the data set the rows 22 and 26 has non valid date formats, so we are going to fix them
#A way that will work for row 26 is to use the method to_datetime
data = pd.read_csv("https://www.w3schools.com/python/pandas/dirtydata.csv.txt")
data["Date"] = pd.to_datetime(data["Date"])
data
```

```python
#now, as we can see the row 26 has now a valid date format, but row 22 has a Not a Time value
#so, one option is to delate that row
data.dropna(subset=["Date"], inplace=True)
data
```

# 3. Wrong data

```python
#Wrong data can be values that are out of the range for a column
#due to typos or any other issue. We can replace these values by using the
#loc[] attribute.
#For instance, the Duration value in row 7 may be wrong, so we would like to change it:
data.loc[7, "Duration" ] = 45
data
```

```python
#or we can use a rule to decide which values to change and ho to do it if there are so many rows:
data = pd.read_csv("https://www.w3schools.com/python/pandas/dirtydata.csv.txt")
data["Date"] = pd.to_datetime(data["Date"])
data.dropna(subset=["Date"], inplace=True)

for cell in data.index: 
    if data.loc[cell, "Duration"]>60: data.loc[cell, "Duration"]=120
data
```

```python
#or if we do not care for those values we can elimite the rows by using the drop() method:
data = pd.read_csv("https://www.w3schools.com/python/pandas/dirtydata.csv.txt")
data["Date"] = pd.to_datetime(data["Date"])
data.dropna(subset=["Date"], inplace=True)

for cell in data.index:
    if data.loc[cell, "Duration"]>60:data.drop(cell, inplace=True)
data
```

# Grouping data

```python
df = pd.DataFrame(
    [
    ("bird", "Falconiformes", 389.0),
    ("bird", "Psittaciformes", 24.0),
    ("mammal", "Carnivora", 80.2),
    ("mammal", "Primates", np.nan),
    ("mammal", "Carnivora", 58),
    ("fish", "Carnivora", 34), 
    ("mammal", "Fruits", 23),
    ("bird", "Omnivoro", 60),
    ("fish", "Fruits", 40)],
    index=["falcon", "parrot", "lion", "monkey", "leopard", "Shrank", "ape", "dove", "whale"], columns=("class", "order", "max_speed"))
df
```

```python
#we start by creating the groups by any criteria
#for instance grouping by "order" will group the df by each 
#of its irders (Falconiformes, Psittaciformes, Carnivora... )
#If we group by ("order", "class") then it will make groups of 
#the cells which have a given order and also share a given class
grouped = df.groupby(by=["order", "class"])
grouped
```

```python
#to see the groups that were created
grouped.groups
```

```python
#the indexes of the elements of each group
grouped.indices
```

```python
#we can see the keys of each group
grouped.groups.keys()
```

```python
#and their elements
grouped.groups.values()
```

```python
#we can see as a df each group
grouped.get_group(("Carnivora", "mammal"))
```

```python
#we can obtain some statistics from the groups
#for intance, the mean
grouped.mean()
```

```python
#the std
grouped.std()
```

```python
#the sum
grouped.sum()
```

```python
#the sum
grouped.sum()["max_speed"]
```

```python
#Or we can know how many elements there are for each group
grouped.count()
```

```python
#the minimun value of each group
grouped.min()
```

```python
grouped.get_group(name=('Carnivora', 'mammal')).value_counts()
```

```python
#we can also aggregate columns to our df (which can be a group) with df.agg()
df3 = pd.DataFrame({"Person": ["John", "Myla", "Lewis", "John", "Myla", "Lewis", "John", "Myla"], 
                       "Age": [24, 55, 24, 21, 26, 54, 64, 55], 
                     "month":[1,1,1,3,3,3,5,5]})
df3
```

```python
df3_group = df3.groupby(by="Person")
df3_group
```

```python
df3_group.groups
```

```python
#for instance, if we want to add the mean to the group
df3_group.agg(func="mean")
```

```python
#or if we want to add the mean and the max value:
df3_group.agg(func=["mean", "max"])
```

```python
#or if we want to add the mean and min values to age and the max to the month:
df3_group.agg(func={"Age":["mean", "min"], "month":"max"})
```
