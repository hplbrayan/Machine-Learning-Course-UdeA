# Ejercicio de Groupby

```python
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(path, sep='|', index_col='user_id')
print(users.shape)
users.head()
```

## ¿Cuántas ocupaciones diferentes hay en este conjunto de datos?

```python
users.occupation.unique(),users.occupation.unique().shape
```

## ¿Cuál es la ocupación más frecuente?

```python
users.occupation.value_counts()
```

```python
users.occupation.mode()
```

## ¿Cuál es la edad media por ocupación?

```python
users.groupby('occupation').mean().sort_values(by='age')
```

## Para cada ocupación calcule las edades mínimas y máximas.

```python
users.groupby('occupation').agg({'age':['max','min']})
```

## Para cada combinación de ocupación y género calcule la edad media.

```python
users.groupby(['occupation','gender']).mean()
```

## Para cada ocupación muestre el porcentaje de mujeres y hombres.

```python
users.groupby(['occupation','gender']).count()
```

```python
users_pivot = users.pivot_table('age',index='occupation',columns='gender',aggfunc='count',margins=True).fillna(0)
```

```python
users_pivot['All']
```

```python
op1 = users_pivot.apply(lambda x: x*100/users_pivot['All'])
op1
```

```python
total_occupation = users.groupby('occupation').count()[['age']]
for_gender = users.groupby(['occupation','gender']).count()[['age']]
op2 = for_gender.div(total_occupation,level='occupation')*100
op2
```

## Descubra la proporción de hombres por ocupación y clasifíquela de mayor a menor

```python
op1['M'].sort_values(ascending=False)
```

```python
op2.loc[[(c,'M') for c in op2.index.get_level_values(0)]]
```

```python
op2.query("gender=='M'").sort_values(by='age',ascending=False)
```
