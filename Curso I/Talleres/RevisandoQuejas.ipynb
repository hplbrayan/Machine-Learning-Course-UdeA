#Revisemos varias quejas

Hagamos en un loop todo lo pedido anteriormente pero para todos los archivos en la carpeta complaints

```python
stop_words_file = open("/content/stop_words_english.txt")
stop_words = stop_words_file.read().split("\n")

informacion = {}
for archivo_elegido in archivos:
  fio_elegido = open("./complaints/" + archivo_elegido,encoding='latin-1')
  texto = fio_elegido.read()
  longitud_archivo_elegido = len(texto)
  texto = texto.replace(".","").replace(",","").replace(";","").lower()
  bolsa_palabras = texto.split(" ")
  bolsa_palabras_limpia = list(set(bolsa_palabras) - set(bolsa_palabras).intersection(stop_words))
  freq_palabras = {p:bolsa_palabras.count(p) for p in bolsa_palabras_limpia}
  informacion[archivo_elegido] = {"longitud":longitud_archivo_elegido,"freq":freq_palabras}
```
